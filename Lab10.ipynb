```python
#importing libraries for image processing
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text=cv2.imread('broken_text.tif',0)
fig_alpha=cv2.imread('Fig01.tif',0)
finger_print=cv2.imread('fp.tif',0)
```

```python
#lab task1
#closing image to remove noise
kernel= cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
fingPrin=cv2.erode(finger_print,kernel,iterations=1)

kernel= cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
#closing to regain some details
fingPrin= cv2.morphologyEx(fingPrin, cv2.MORPH_CLOSE, kernel,iterations=2)

#using elliptical structuring element
kernel= cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
#closing to regain some details
fingPrin= cv2.morphologyEx(fingPrin, cv2.MORPH_CLOSE, kernel)

#displaying output
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(finger_print,cmap='gray')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(fingPrin,cmap='gray')
plt.title('Eroded Image')
```

```python
#lab task 2
#initializing rect structuring element
kernel= cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
#dilating image
ntext= cv2.dilate(text,kernel,iterations=2) 

#initializing cross structuring element
kernel= cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
#eroding image
ntext= cv2.erode(ntext,kernel,iterations=2)

# kernel= cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))
# #eroding image
# ntext= cv2.erode(ntext,kernel,iterations=2)



#displaying output
plt.figure(figsize=(20,14))
plt.subplot(1,2,1)
plt.imshow(text,cmap='gray')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow( ntext,cmap='gray')
plt.title('Dilated Image')
```
